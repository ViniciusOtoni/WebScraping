```python
from sklearn.pipeline import Pipeline
import ETL 
import pandas as pd
import warnings
```

```python
df = pd.read_csv('../data/d01_raw/books_data.csv')
pd.set_option('display.max_columns', None)
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
str_pipe = Pipeline(steps=[
    ("ConvertPriceToNumber", ETL.ConvertPriceToNumber(column_name="Book_Price")),
    ("TransformPriceIntoReal", ETL.TransformPriceIntoReal(column_name="Book_Price")),
    ("ConvertRatingToNumber", ETL.ConvertRatingToNumber(column_name="Book_Rating")),
    ("QuantityBookAvailable", ETL.QuantityBookAvailable(column_name="Book_Stock")),
    ("TransformBookStock", ETL.TransformBookStock(column_name="Book_Stock")),
    ("ChangeOrderColumns", ETL.ChangeOrderColumns(column_names=["Book_Category", "Book_Title", "Book_Price", "Book_Price_BR", "Book_Stock", "Book_Stock_Available", "Book_Rating"]))
]) 
```

```python
example = str_pipe.fit_transform(df)
example.head(15)
```

```python
example.to_csv("../data/d02_intermediate/books_data_cleaned.csv", index=False)
```
